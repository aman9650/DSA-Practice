```python
# we can put or remove null value according to different price of different category of products
data['Category'].value_counts()
```

```python

data.groupby('Category').mean()['DiscountPrice (in Rs)']
```

```python
#in the following method, we have used groupby() and transform() functions to replace Null values.
data['DiscountPrice (in Rs)'] = data['DiscountPrice (in Rs)'].fillna(data.groupby('Category')['DiscountPrice (in Rs)'].transform('mean'))
```

```python
data['DiscountPrice (in Rs)'].describe() 
# now we can see that the mean value changes
```
